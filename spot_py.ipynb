O pr√≥ximo passo √© construir de fato o sistema de recomenda√ß√£o, transformando a Matriz de Conte√∫do (m√∫sica e caracter√≠sticas) em n√∫meros que o sistema pode usar para calcular a relev√¢ncia de cada termo, e depois, encontrar as m√∫sicas mais parecidas com o perfil de um novo usu√°rio.

Para fazer esse c√°lculo ser√£o necess√°rias a biblioteca scikit-learn:

```python
#instala a biblioteca scikit-learn
pip install scikit-learn

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel # fun√ß√£o para Similaridade do Cosseno

#carrega a Matriz de Conte√∫do
matriz_conteudo = pd.read_csv('matriz_conteudo.csv')

print("Matriz de Conte√∫do Carregada com Sucesso.")
```

Para utilizar essa biblioteca √© necess√°rio que os dados de cada estejam em uma √∫nica string, dessa forma, temos que agrupar as colunas da matriz de conte√∫do.

```python
#combina as caracter√≠sticas em uma √∫nica coluna de texto
def combinar_caracteristicas(row):
    # para que o TF-IDF trate cada termo (ex: "Pop" e "R&B") como separado.
    caracteristicas = f"{row['G√™nero']} {row['Artista']} {row['Humor']} {row['√âpoca']}"
    return caracteristicas.replace(',', '')
    #substitui as v√≠rgulas por ' "

#aplica a fun√ß√£o para criar a nova coluna 'tags_conteudo'
matriz_conteudo['tags_conteudo'] = matriz_conteudo.apply(combinar_caracteristicas, axis=1)

print("\nExemplo da nova coluna de tags:")
print(matriz_conteudo[['Nome da M√∫sica', 'tags_conteudo']].head())
```

Agora que os dados est√£o organizados, a fun√ß√£o TfidfVectorizer calcula os pesos, diminuindo os peso de palavras muito comuns, aumentando o de termos mais espec√≠ficos.

```python
tfidf = TfidfVectorizer(stop_words='english')

#transforma a coluna de tags na matriz de pesos TF-IDF
matriz_tfidf = tfidf.fit_transform(matriz_conteudo['tags_conteudo'])

#converte o resultado para uma matriz
matriz_tfidf_df = pd.DataFrame(matriz_tfidf.toarray(), columns=tfidf.get_feature_names_out())

print("\nMatriz TF-IDF (M√∫sicas x Termos de Conte√∫do):")
#exibe as primeiras 5 linhas e 10 colunas
print(matriz_tfidf_df.iloc[:5, :10])
```

Agora √© necess√°rio comparar a matriz TF-IDF para criar a Matriz de Similaridade. O resultado ser√° uma tabela 24x24 (m√∫sicas x m√∫sicas) que diz o qu√£o parecidas elas s√£o. Quanto mais pr√≥ximo de 1, mais similares s√£o as m√∫sicas.

```python
#a Similaridade do Cosseno calcula o √¢ngulo entre os vetores (perfis) das m√∫sicas.
matriz_similaridade_cosseno = linear_kernel(matriz_tfidf, matriz_tfidf)

print("\nMatriz de Similaridade do Cosseno (M√∫sicas x M√∫sicas):")
# exibe a matriz de similaridade (a diagonal ser√° 1.0, pois a m√∫sica √© 100% parecida consigo mesma)
print(pd.DataFrame(matriz_similaridade_cosseno).iloc[:5, :5])
```

Com o sistema praticamente conclu√≠do, √© hora de construir a fun√ß√£o de recomenda√ß√£o. Para isso precisamos mapear cada m√∫sica baseada em seu √≠ndice.

```python
#utiliza o √≠ndice da Matriz de Conte√∫do para mapear o nome das m√∫sicas
indices = pd.Series(matriz_conteudo.index, index=matriz_conteudo['Nome da M√∫sica']).drop_duplicates()
```

No pr√≥ximo passo √© criada ativamente fun√ß√£o que usa a Matriz de Similaridade. Nessa fun√ß√£o as prefer√™ncias do usu√°rio s√£o transformadas em uma √∫nica string e utilizamos o vetorizador, assim como na matriz de conte√∫do.

```python
from sklearn.metrics.pairwise import linear_kernel

def get_recommendations_by_profile(user_preferences_string, tfidf_vectorizer, matriz_musicas, matriz_tfidf_musicas, n=3):

    #cria o Vetor do Usu√°rio
    vetor_usuario = tfidf_vectorizer.transform([user_preferences_string])

    #calcula a Similaridade (usu√°rio vs. m√∫sicas)
    similaridade_cosseno = linear_kernel(vetor_usuario, matriz_tfidf_musicas).flatten()

    #encontra as melhores recomenda√ß√µes
    indices_recomendados = similaridade_cosseno.argsort()[::-1]
    melhores_indices = indices_recomendados[:n]

    #retorna nome da M√∫sica e artista
    return matriz_musicas[['Nome da M√∫sica', 'Artista']].iloc[melhores_indices]

# --- Teste da Fun√ß√£o ---
preferencias_teste = "Hip-Hop Reflexivo Anos 10"
recomendacoes_df = get_recommendations_by_profile(
    preferencias_teste, 
    tfidf,
    matriz_conteudo,
    matriz_tfidf,
    n=3
)
print("Teste")
print(recomendacoes_df)
```

O pr√≥ximo passo √© criar a integra√ß√£o entre os inputs do usu√°rio e a fun√ß√£o de recomenda√ß√£o, exibindo o resultado da busca.

```python
from IPython.display import display, HTML
import ipywidgets as widgets

def callback_recomendacao(genero, humor, epoca):
    #combina as prefer√™ncias do usu√°rio
    user_preferences_string = f"{genero} {humor} {epoca}"
    
    #chama a fun√ß√£o de recomenda√ß√£o, que agora retorna um DataFrame
    recomendacoes_df = get_recommendations_by_profile(
        user_preferences_string, 
        tfidf, 
        matriz_conteudo, 
        matriz_tfidf, 
        n=3
    )
    
    #formata e exibe o resultado
    html_output = "<h3>üé∂ Suas 3 M√∫sicas Recomendadas:</h3>"
    html_output += "<ol>" 
    
    #itera sobre as linhas do DataFrame para pegar M√∫sica e Artista
    for index, row in recomendacoes_df.iterrows():
        musica = row['Nome da M√∫sica']
        artista = row['Artista']
        html_output += f"<li><strong>{musica}</strong> (Artista: {artista})</li>"
    
    html_output += "</ol>"
    
    #exibe a lista de recomenda√ß√µes
    display(HTML(html_output))
```

E para finalizar, a parte visual e interativa, que recebe os inputs do usu√°rio. Um pequeno formu√°rio, com respostas pr√© definidas para que o usu√°rio possa selecionar suas prefer√™ncias e obter as m√∫sicas que mais se encaixam no seu perfil.

```python
#define as op√ß√µes de escolha para o usu√°rio

opcoes_genero = ['Hip-Hop', 'Pop', 'Funk', 'Samba', 'Pagode', 'Trap', 'Alternativo', 'Rock and Roll', 'MPB']
opcoes_humor = ['Animado', 'Reflexivo', 'Melanc√≥lico', 'Dram√°tico', 'Rom√¢ntico', 'Confiante', 'Nost√°lgico']
opcoes_epoca = ['Anos 20', 'Anos 10', 'Anos 00', 'Anos 90', 'Anos 80', 'Anos 70', 'Anos 60']

style = {'description_width': 'initial'} 

#criar os widgets de sele√ß√£o, para cada uma das perguntas(genero, humor, epoca)
genero_widget = widgets.Dropdown(
    options=opcoes_genero,
    value='Pop',
    description='G√™nero Favorito:',
    disabled=False,
    style=style
)

humor_widget = widgets.Dropdown(
    options=opcoes_humor,
    value='Animado',
    description='Humor Desejado:',
    disabled=False,
    style=style 
)

epoca_widget = widgets.Dropdown(
    options=opcoes_epoca,
    value='Anos 10',
    description='√âpoca Preferida:',
    disabled=False,
    style=style 
)


# criar o bot√£o de recomenda√ß√£o
botao_recomendar = widgets.Button(description="Receber Recomenda√ß√µes!")
output = widgets.Output()
# cria a fun√ß√£o que ser√° chamada quando o bot√£o for clicado
def on_button_clicked(b):
    with output:
        output.clear_output() # Limpa o resultado anterior
        #chama a fun√ß√£o de callback com os valores atuais dos dropdowns
        callback_recomendacao(genero_widget.value, humor_widget.value, epoca_widget.value)

#ligar o clique do bot√£o √† fun√ß√£o
botao_recomendar.on_click(on_button_clicked)

#exibe a interface
print("üéß Spot.py: Sistema de Recomenda√ß√£o por Conte√∫do")
print("‚≠ê Projete seu Perfil e Receba 3 Recomenda√ß√µes:")
display(widgets.VBox([genero_widget, humor_widget, epoca_widget, botao_recomendar, output]))
```
