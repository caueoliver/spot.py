O próximo passo é construir de fato o sistema de recomendação, transformando a Matriz de Conteúdo (música e características) em números que o sistema pode usar para calcular a relevância de cada termo, e depois, encontrar as músicas mais parecidas com o perfil de um novo usuário.

Para fazer esse cálculo serão necessárias a biblioteca scikit-learn:

```python
#instala a biblioteca scikit-learn
pip install scikit-learn

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel # função para Similaridade do Cosseno

#carrega a Matriz de Conteúdo
matriz_conteudo = pd.read_csv('matriz_conteudo.csv')

print("Matriz de Conteúdo Carregada com Sucesso.")
```

Para utilizar essa biblioteca é necessário que os dados de cada estejam em uma única string, dessa forma, temos que agrupar as colunas da matriz de conteúdo.

```python
#combina as características em uma única coluna de texto
def combinar_caracteristicas(row):
    # para que o TF-IDF trate cada termo (ex: "Pop" e "R&B") como separado.
    caracteristicas = f"{row['Gênero']} {row['Artista']} {row['Humor']} {row['Época']}"
    return caracteristicas.replace(',', '')
    #substitui as vírgulas por ' "

#aplica a função para criar a nova coluna 'tags_conteudo'
matriz_conteudo['tags_conteudo'] = matriz_conteudo.apply(combinar_caracteristicas, axis=1)

print("\nExemplo da nova coluna de tags:")
print(matriz_conteudo[['Nome da Música', 'tags_conteudo']].head())
```

Agora que os dados estão organizados, a função TfidfVectorizer calcula os pesos, diminuindo os peso de palavras muito comuns, aumentando o de termos mais específicos.

```python
tfidf = TfidfVectorizer(stop_words='english')

#transforma a coluna de tags na matriz de pesos TF-IDF
matriz_tfidf = tfidf.fit_transform(matriz_conteudo['tags_conteudo'])

#converte o resultado para uma matriz
matriz_tfidf_df = pd.DataFrame(matriz_tfidf.toarray(), columns=tfidf.get_feature_names_out())

print("\nMatriz TF-IDF (Músicas x Termos de Conteúdo):")
#exibe as primeiras 5 linhas e 10 colunas
print(matriz_tfidf_df.iloc[:5, :10])
```

Agora é necessário comparar a matriz TF-IDF para criar a Matriz de Similaridade. O resultado será uma tabela 24x24 (músicas x músicas) que diz o quão parecidas elas são. Quanto mais próximo de 1, mais similares são as músicas.

```python
#a Similaridade do Cosseno calcula o ângulo entre os vetores (perfis) das músicas.
matriz_similaridade_cosseno = linear_kernel(matriz_tfidf, matriz_tfidf)

print("\nMatriz de Similaridade do Cosseno (Músicas x Músicas):")
# exibe a matriz de similaridade (a diagonal será 1.0, pois a música é 100% parecida consigo mesma)
print(pd.DataFrame(matriz_similaridade_cosseno).iloc[:5, :5])
```

Com o sistema praticamente concluído, é hora de construir a função de recomendação. Para isso precisamos mapear cada música baseada em seu índice.

```python
#utiliza o índice da Matriz de Conteúdo para mapear o nome das músicas
indices = pd.Series(matriz_conteudo.index, index=matriz_conteudo['Nome da Música']).drop_duplicates()
```

No próximo passo é criada ativamente função que usa a Matriz de Similaridade. Nessa função as preferências do usuário são transformadas em uma única string e utilizamos o vetorizador, assim como na matriz de conteúdo.

```python
from sklearn.metrics.pairwise import linear_kernel

def get_recommendations_by_profile(user_preferences_string, tfidf_vectorizer, matriz_musicas, matriz_tfidf_musicas, n=3):

    #cria o Vetor do Usuário
    vetor_usuario = tfidf_vectorizer.transform([user_preferences_string])

    #calcula a Similaridade (usuário vs. músicas)
    similaridade_cosseno = linear_kernel(vetor_usuario, matriz_tfidf_musicas).flatten()

    #encontra as melhores recomendações
    indices_recomendados = similaridade_cosseno.argsort()[::-1]
    melhores_indices = indices_recomendados[:n]

    #retorna nome da Música e artista
    return matriz_musicas[['Nome da Música', 'Artista']].iloc[melhores_indices]

# --- Teste da Função ---
preferencias_teste = "Hip-Hop Reflexivo Anos 10"
recomendacoes_df = get_recommendations_by_profile(
    preferencias_teste, 
    tfidf,
    matriz_conteudo,
    matriz_tfidf,
    n=3
)
print("Teste")
print(recomendacoes_df)
```

O próximo passo é criar a integração entre os inputs do usuário e a função de recomendação, exibindo o resultado da busca.

```python
from IPython.display import display, HTML
import ipywidgets as widgets

def callback_recomendacao(genero, humor, epoca):
    #combina as preferências do usuário
    user_preferences_string = f"{genero} {humor} {epoca}"
    
    #chama a função de recomendação, que agora retorna um DataFrame
    recomendacoes_df = get_recommendations_by_profile(
        user_preferences_string, 
        tfidf, 
        matriz_conteudo, 
        matriz_tfidf, 
        n=3
    )
    
    #formata e exibe o resultado
    html_output = "<h3>🎶 Suas 3 Músicas Recomendadas:</h3>"
    html_output += "<ol>" 
    
    #itera sobre as linhas do DataFrame para pegar Música e Artista
    for index, row in recomendacoes_df.iterrows():
        musica = row['Nome da Música']
        artista = row['Artista']
        html_output += f"<li><strong>{musica}</strong> (Artista: {artista})</li>"
    
    html_output += "</ol>"
    
    #exibe a lista de recomendações
    display(HTML(html_output))
```

E para finalizar, a parte visual e interativa, que recebe os inputs do usuário. Um pequeno formuário, com respostas pré definidas para que o usuário possa selecionar suas preferências e obter as músicas que mais se encaixam no seu perfil.

```python
#define as opções de escolha para o usuário

opcoes_genero = ['Hip-Hop', 'Pop', 'Funk', 'Samba', 'Pagode', 'Trap', 'Alternativo', 'Rock and Roll', 'MPB']
opcoes_humor = ['Animado', 'Reflexivo', 'Melancólico', 'Dramático', 'Romântico', 'Confiante', 'Nostálgico']
opcoes_epoca = ['Anos 20', 'Anos 10', 'Anos 00', 'Anos 90', 'Anos 80', 'Anos 70', 'Anos 60']

style = {'description_width': 'initial'} 

#criar os widgets de seleção, para cada uma das perguntas(genero, humor, epoca)
genero_widget = widgets.Dropdown(
    options=opcoes_genero,
    value='Pop',
    description='Gênero Favorito:',
    disabled=False,
    style=style
)

humor_widget = widgets.Dropdown(
    options=opcoes_humor,
    value='Animado',
    description='Humor Desejado:',
    disabled=False,
    style=style 
)

epoca_widget = widgets.Dropdown(
    options=opcoes_epoca,
    value='Anos 10',
    description='Época Preferida:',
    disabled=False,
    style=style 
)


# criar o botão de recomendação
botao_recomendar = widgets.Button(description="Receber Recomendações!")
output = widgets.Output()
# cria a função que será chamada quando o botão for clicado
def on_button_clicked(b):
    with output:
        output.clear_output() # Limpa o resultado anterior
        #chama a função de callback com os valores atuais dos dropdowns
        callback_recomendacao(genero_widget.value, humor_widget.value, epoca_widget.value)

#ligar o clique do botão à função
botao_recomendar.on_click(on_button_clicked)

#exibe a interface
print("🎧 Spot.py: Sistema de Recomendação por Conteúdo")
print("⭐ Projete seu Perfil e Receba 3 Recomendações:")
display(widgets.VBox([genero_widget, humor_widget, epoca_widget, botao_recomendar, output]))
```
